## 1.1 `if / elif / else`
En esta clase aprendí a hacer que Python tome decisiones usando condiciones.
Sirve para ejecutar código **solo si** se cumple una condición.

**Ideas clave:**
- La condición debe ser algo que Python pueda interpretar como `True` o `False`.
- La indentación (espacios) define el bloque de código.

````python
#Defino una nota y uso if/elif/else para decidir qué mensaje mostrar
nota = 15
if nota > 15:
  print("muy bien")
elif nota > 10:
  print("aprobado")
else:
  print("desaprobado")
````

````python
#Recorro una lista de frutas y en cada vuelta imprimo el texto "fruta:" junto al elemento.
frutas = ["manzana", "pera", "uva"]

for f in frutas:
    print("fruta:", f)
````

````python
#Uso enumerate para recorrer la lista y además obtener un índice (contador) empezando en 1.
for i, fruta in enumerate(frutas, start=1):
    print(i, "->", fruta)
````

````python
#Igual que arriba, pero ahora el índice empieza en 2 (start cambia el número inicial del contador).
for i, fruta in enumerate(frutas, start=2):
    print(i, "->", fruta)
````

## 1.3 `while`: repetir mientras una condición sea verdadera
Se usa cuando **no sabes cuántas iteraciones** necesitas.

**Cuidado:** si la condición nunca cambia → loop infinito.

````python
# Ejemplo de while: repite un bloque de código mientras la condición sea True.
# Aquí usamos un contador que empieza en 0 y el ciclo corre mientras contador < 10.
# En cada vuelta imprimimos el valor actual y luego lo aumentamos en 1 con contador += 1.
# Si no aumentáramos el contador, la condición nunca cambiaría y el loop podría volverse infinito.
contador = 0
while contador < 10:
  print(contador)
  contador += 1 #contador = contador + 1
````

````python
# Ejemplo del operador += (atajo para sumar y guardar el resultado en la misma variable).
# numero += 2 significa lo mismo que numero = numero + 2.
# Sirve para actualizar variables de forma más corta y es muy común dentro de loops.
numero = 9
numero += 2 # numero = numero + 2
print(numero)
````

## 1.4 `break` y `continue`
- `break`: sale del loop
- `continue`: salta a la siguiente iteración

````python
#Uso break para salir del for cuando i pasa de 5 (se detiene el loop).
for i in range(10):
  if i > 5:
    break
  print(i)
````

````python
#Imprimo solo los números impares usando i % 2 == 1 (módulo = residuo).
for i in range(10):
  if i % 2 == 1:
    print(i)
````

````python
#Combino dos condiciones: solo impares y corto el loop cuando i > 5 con break.
for i in range(10):
  if i % 2 == 1:
    if i > 5:
      break
    print(i)
````

````python
#Uso continue para saltar los pares (i % 2 == 0) y seguir con la siguiente iteración.
for i in range(10):
  if i % 2 == 0:
    continue
  print(i)
````

````python
# Uso while para recorrer contador de 0 a 9 y saltar los pares con continue.
# Ojo: si uso continue debo asegurar que contador aumente antes, para no quedarme en loop infinito.
contador = 0
while contador < 10:
  if contador % 2 == 0:
    #contador += 1
    continue
  print(contador)
  contador += 1
````

# 2) Ejercicio: “filtrar + contar”

Vamos a simular datos de pedidos (lista de diccionarios) y:
- quedarnos con pedidos grandes,
- contar por ciudad,
- obtener un resumen.


````python
# Creo una lista llamada orders donde cada elemento es un diccionario con datos de un pedido.
# Cada pedido tiene id, user_id, ciudad y monto; esto servirá para filtrar y contar después.
orders = [
    {"order_id": 1, "user_id": 10, "city": "Lima", "amount": 120.0},
    {"order_id": 2, "user_id": 11, "city": "Cusco", "amount": 35.5},
    {"order_id": 3, "user_id": 10, "city": "Lima", "amount": 80.0},
    {"order_id": 4, "user_id": 12, "city": "Arequipa", "amount": 220.0},
    {"order_id": 5, "user_id": 13, "city": "Cusco", "amount": 150.0},
]
````

````python
print(type(orders))
````

````python
# Recorro la lista orders y filtro solo los pedidos cuyo monto (amount) es mayor a 100.
# Los pedidos que cumplen la condición los guardo en big_orders para analizarlos después.
big_orders = []
for o in orders:
    if o["amount"] >= 100:
        big_orders.append(o)

print(big_orders)
````

````python
# Calculo cuántos pedidos quedaron luego del filtro usando len(big_orders).
# Esto me sirve para verificar rápidamente si el filtro funcionó como esperaba.

print(len(big_orders))
````

````python
# Creo un diccionario counts para contar cuántos pedidos grandes hay por ciudad.
# Si la ciudad no existe aún en counts la inicializo en 0 y luego voy sumando 1 por pedido.
# contar por ciudad
counts = {}

for o in big_orders:
    city = o["city"]
    if city not in counts:
        counts[city] = 0
    counts[city] += 1

counts
````

## 3.1 Leer un archivo de texto con `with open(...)`
`with` asegura que el archivo se cierra bien (incluso si algo falla).

````python
## Guardo datos en formato tipo CSV dentro de un string multilínea (cada fila es un pedido).
# Esto sirve para simular un archivo de texto y practicar lectura/escritura con open().
text = """order_id,user_id,city,amount
1,10,Lima,120.0
2,11,Cusco,35.5
3,10,Lima,80.0
4,12,Arequipa,220.0
5,13,Cusco,150.0
"""
````

````python
# Verifico que text sea un string (str) y cuántos caracteres contiene.
# Esto me ayuda a confirmar que el “contenido del archivo” está bien armado antes de guardarlo.
print(type(text))
print(len(text))
````

````python
# Creo (o reemplazo) el archivo orders.csv en modo escritura ("w") y guardo el texto dentro.
# Uso encoding="utf-8" para evitar problemas con caracteres especiales.
with open("orders.csv", "w", encoding="utf-8") as f:
    f.write(text)
````

````python
# Abro el archivo en modo lectura ("r") y lo leo completo con f.read().
# Guardo todo el contenido en la variable content y luego lo imprimo para comprobarlo.
with open("orders.csv", "r", encoding="utf-8") as f:
    content = f.read()
print(content)
````

````python
# Verifico el tipo de content: al leer con read() obtengo un string (str).
# Aún no es una lista ni una tabla, solo texto completo.
print(type(content))
````

````python
# que esta pasando?
# Aquí no leo todo de golpe: recorro el archivo línea por línea con for line in f.
# strip() quita el salto de línea al final y así imprimo cada fila limpia con "Linea:".
with open("orders.csv", "r", encoding="utf-8") as f:
    for line in f:
        line = line.strip()
        print("Linea:", line)
````

````python
# Ejemplo 1: Limpio líneas tipo CSV, valido datos y hago un resumen por ciudad.
# Uso control de flujo (if/continue) para saltar filas inválidas sin romper el programa.

raw_lines = [
    "order_id,user_id,city,amount",
    "1,10,Lima,120.0",
    "2,11,Cusco,35.5",
    "",                                # línea vacía
    "3,10,Lima,80.0",
    "4,12,Arequipa,220.0",
    "5,13,Cusco,150.0",
    "6,14,Lima,NO_NUM",                # monto inválido
    "7,15,,50.0",                      # ciudad vacía
    "8,16,Trujillo,-10",               # monto negativo (lo considero inválido)
]

orders_clean = []
invalid_count = 0

# Recorro desde la segunda línea (me salto el header)
for line in raw_lines[1:]:
    line = line.strip()

    # 1) salto líneas vacías
    if line == "":
        invalid_count += 1
        continue

    parts = line.split(",")

    # 2) valido que tenga 4 campos
    if len(parts) != 4:
        invalid_count += 1
        continue

    order_id, user_id, city, amount = parts
    city = city.strip()

    # 3) valido ciudad no vacía
    if city == "":
        invalid_count += 1
        continue

    # 4) valido monto numérico y positivo
    try:
        amount = float(amount)
    except ValueError:
        invalid_count += 1
        continue

    if amount <= 0:
        invalid_count += 1
        continue

    orders_clean.append({
        "order_id": int(order_id),
        "user_id": int(user_id),
        "city": city,
        "amount": amount
    })

print("Pedidos válidos:", len(orders_clean))
print("Pedidos inválidos:", invalid_count)
print("Ejemplo de pedido limpio:", orders_clean[0])
````
